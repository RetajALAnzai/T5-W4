## Preprocessing

### Drop unnecessary columns

```python
df.columns
```

```python
df.drop(columns=["TotalCharges"],inplace=True)
```

### One-hot-encoder and Label-encoder

```python
encode_cols=["gender","Partner","Dependents","PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies","Contract","PaperlessBilling","PaymentMethod"]
ohe_encode = OneHotEncoder(sparse_output=False)

one_hot_encoded = pd.DataFrame(ohe_encode.fit_transform(df[encode_cols]))

one_hot_df = pd.DataFrame(one_hot_encoded, columns=ohe_encode.get_feature_names_out(encode_cols))
df = pd.concat([df, one_hot_encoded], axis=1)

df_encoded = df

encoding_columns = encode_cols

df_encoded = df_encoded.drop(encoding_columns, axis=1)
df = df_encoded
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        label_encoder = LabelEncoder()
        df[column] = label_encoder.fit_transform(df[column])
```

## Selecting Features && Train Test Split

### Selecting Features

```python
x=df.drop(columns=["Churn"])
y=df["Churn"]
```

### Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
for column in x_train.columns:
    if x_train[column].dtype == 'object':
        label_encoder = LabelEncoder()
        x_train[column] = label_encoder.fit_transform(x_train[column])
```

## Modeling

### Creating the sequential model

```python
model= Sequential()
```

### Adding the input layer

```python
model.add(Dense(units=6, activation='relu'))
```

### Adding the hidden layers

```python
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=6, activation='relu'))
```

### Adding the output layer

```python
model.add(Dense(units=1, activation='sigmoid'))
```

### Compiling the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Display model summary

```python
model.summary()
```

### Fitting the model

```python
model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

## Evaluating the model

```python
model.evaluate(x_test,y_test)
```
